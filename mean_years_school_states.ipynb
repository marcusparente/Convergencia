```python
mean_years_school_states = mean_years_school_states.dropna(how='any')
mean_years_school_states.tail()
```

```python
# Filtrar linhas onde a coluna 'coluna' contenha 'RM_RIDE'
mean_years_school_states = mean_years_school_states[~mean_years_school_states['AGREGACAO'].str.contains('RM_RIDE')]
```

```python
mean_years_school_states.tail()
```

```python
mean_years_school_states['CODIGO'] = mean_years_school_states['CODIGO'].astype(int)
```

```python
mean_years_school_states.info()
```

```python
mean_years_school_states.columns
```

```python
mean_years_school_states = mean_years_school_states.drop(columns=['AGREGACAO', 'NOME'])
```

```python
mean_years_school_states.head(2)
```

```python
mean_years_school_states.rename(columns={'ANO': 'time'}, inplace=True)
```

```python
mean_years_school_states.head(2)
```

```python
# Reorganizando o DataFrame para ter os estados como colunas e datas como índices
mean_years_school_states = mean_years_school_states.pivot(index='time', columns='CODIGO')['ANOSEST']

# Removendo o nome da coluna de estados ('Código')
mean_years_school_states.columns.name = None
```

```python
mean_years_school_states.reset_index(inplace=True)
```
