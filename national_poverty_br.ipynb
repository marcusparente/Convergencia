```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from bs4 import BeautifulSoup
import requests
import warnings
warnings.filterwarnings('ignore')
```

```python
#Carregando nosso DF
os.chdir('/content/drive/MyDrive/Bases de Dados para Indicadores EUROFOUND/Base de Dados')
national_poverty_br = pd.read_csv('taxa_pobreza_BR_nacional.csv', sep=';')
```

```python
print(national_poverty_br)
```

```python
#Selecionando as colunas do DF que precisamos
national_poverty_br = national_poverty_br.drop(['Sigla', 'Brasil'], axis=1)
national_poverty_br.head()
```

```python
#Transpondo nosso DF
national_poverty_br_T = national_poverty_br.T
national_poverty_br_T.head()
```

```python
#Criando o index
national_poverty_br_T = national_poverty_br_T.reset_index(drop=False)
national_poverty_br_T.head()
```

```python
#Renomeando nosso index e removendo a primeira linha repetida
new_columns_names = national_poverty_br_T.iloc[0].tolist()
new_columns_names= dict(zip(national_poverty_br_T.columns, new_columns_names))
national_poverty_br_T.rename(columns = new_columns_names, inplace=True)
print(national_poverty_br_T)
```

```python
national_poverty_br_T = national_poverty_br_T.drop(index=0)
print(national_poverty_br_T)
```

```python
#Renomeando o nome das colunas
national_poverty_br_T.rename(columns = {'Código' : 'time'}, inplace = True)
national_poverty_br_T.rename(columns = {0 : 'BR'}, inplace = True)
print(national_poverty_br_T)
```

```python
#Salvando nosso DF
national_poverty_br = national_poverty_br_T
national_poverty_br.to_csv('/content/drive/MyDrive/Bases de Dados para Indicadores EUROFOUND/Bases Finais/national_poverty_br.csv',encoding='UTF-8', index = False)
```
