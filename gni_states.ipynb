# PIB estadual

### fonte: ipeadata (http://www.ipeadata.gov.br/Default.aspx)

Pesquisar PIB Estadual - preços de mercado (preços de 2010)

Escolher a desagregação "Estados"

No canto superior direito exportar arquivo em csv com separador ";"

```python
ipeadata_pib = pd.read_csv('data\pib_estadual_valores_2010.csv', delimiter=';', skiprows=1)
```

```python
ipeadata_pib.head()
```

```python
ipeadata_pib.columns
```

```python
ipeadata_pib = ipeadata_pib.drop(columns = ['Sigla', 'Estado','Unnamed: 67'])
```

```python
ipeadata_pib.rename(columns={'Código': 'codigo'}, inplace=True)
```

```python
ipeadata_pib.head()
```

```python
gni_states_br = ipeadata_pib.melt(id_vars='codigo', var_name='Ano', value_name='pib')
```

```python
gni_states_br.head()
```

```python
# Reorganizando o DataFrame para ter os estados como colunas e datas como índices
gni_states_br = gni_states_br.pivot(index='Ano', columns='codigo')['pib']

# Removendo o nome da coluna de estados ('Código')
gni_states_br.columns.name = None
```

```python
gni_states_br.head()
```

```python
gni_states_br.reset_index(inplace=True)
```

```python
gni_states_br.tail()
```

```python
gni_states_br.rename(columns={'Ano': 'time'}, inplace=True)
```

```python
gni_states_br.head()
```

```python
gni_states_br = gni_states_br.dropna()
```

```python
gni_states_br.head()
```

```python
gni_states_br.tail()
```

```python
gni_states_br.columns
```

```python
gni_states_br.info()
```

```python
gni_states_br['time'] = gni_states_br['time'].astype(int)
```

```python
gni_states_br.info()
```

```python
# Converter colunas de estados de object para float, substituindo vírgulas por pontos
def convert_to_float(x):
    if isinstance(x, str):
        try:
            return float(x.replace(',', '.'))
        except ValueError:
            return x
    return x

# Aplicar a função convert_to_float em todas as células do DataFrame
gni_states_br = gni_states_br.applymap(convert_to_float)
```
