```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from bs4 import BeautifulSoup
import requests
import warnings
warnings.filterwarnings('ignore')
```

```python
#Carregando nosso DF
os.chdir('/content/drive/MyDrive/Bases de Dados para Indicadores EUROFOUND/Base de Dados')
homicide_rate_women_states = pd.read_csv('taxa-de-homicidios-mulheres.csv', sep = ';')
```

```python
homicide_rate_women_states.head()
```

```python
#Selecionando as colunas que precisamos
homicide_rate_women_states = homicide_rate_women_states.drop(['nome'], axis = 1)
homicide_rate_women_states.head()
```

```python
#Pivotando nosso dataframe
homicide_rate_women_states_T = homicide_rate_women_states.pivot_table(index = 'período', columns = 'cod', values = 'valor')
homicide_rate_women_states_T
```

```python
homidicide_rate_women_states_T = homicide_rate_women_states_T.reset_index(inplace = True)
print(homicide_rate_women_states_T)
```

```python
homicide_rate_women_states_T = homicide_rate_women_states_T.rename(columns = {'período': 'time'})
homicide_rate_women_states_T.head()
```

```python
#Salvando nosso DF
homicide_rate_women_states = homicide_rate_women_states_T
homicide_rate_women_states.to_csv('/content/drive/MyDrive/Bases de Dados para Indicadores EUROFOUND/Bases Finais/homicide_rate_women_states.csv', encoding='UTF-8', index = False)
```
