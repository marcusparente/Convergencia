```python
women_management_positions_T = women_management_positions.T
print(women_management_positions_T)
```

```python
women_management_positions_T = women_management_positions_T.reset_index()
women_management_positions_T.head()
```

```python
#Renomeando nosso index e removendo a primeira linha repetida
new_columns_names = women_management_positions_T.iloc[0].tolist()
new_columns_names= dict(zip(women_management_positions_T.columns, new_columns_names))
women_management_positions_T.rename(columns = new_columns_names, inplace=True)
print(women_management_positions_T)
```

```python
women_management_positions_T = women_management_positions_T.drop(index=0)
print(women_management_positions_T)
```

```python
women_management_positions_T.rename(columns = {'Acre' : '12', 'Alagoas' : '27', 'Amazonas' : '13', 'Amapá' : '16', 'Bahia' : '29', 'Ceará' : '23', 'Distrito Federal' : '53', 'Espírito Santo' : '32', 'Goiás' : '52', 'Maranhão' : '21', "Minas Gerais" : '31', 'Mato Grosso do Sul' : '50', 'Mato Grosso' : '51', 'Pará' : '15', 'Paraíba' : '25', 'Pernambuco' : '26', 'Piauí' : '22', 'Paraná' : '41', 'Rio de Janeiro' : '33', 'Rio Grande do Norte' : '24', 'Rondônia' : '11', 'Roraima' : '14', 'Rio Grande do Sul' : '43', 'Santa Catarina' : '42', 'Sergipe' : '28', 'São Paulo' : '35', 'Tocantins' : '17'}, inplace = True)
print(women_management_positions_T)
```

```python
women_management_positions_T = women_management_positions_T.rename(columns = {'UF' : 'time'})
print(women_management_positions_T)
```
