```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from bs4 import BeautifulSoup
import requests
import warnings
warnings.filterwarnings('ignore')
```

```python
#Carregando nosso DF
national_poverty_states_br = pd.read_csv('taxa_pobreza_nacional_estados.csv', sep=';')
```

```python
print(national_poverty_states_br)
```

```python
national_poverty_states_br.head()
```

```python
#Selecionando as colunas que precisamos
national_poverty_states_br = national_poverty_states_br.drop(['Sigla', 'Estado'], axis=1)
national_poverty_states_br.head()
```

```python
#Removendo a última coluna que está vazia
national_poverty_states_br = national_poverty_states_br.iloc[: , :-1]
print(national_poverty_states_br)
```

```python
#Transpondo nosso DF
national_poverty_states_br_T = national_poverty_states_br.T
print(national_poverty_states_br_T)
```

```python
#Criando o index
national_poverty_states_br_T = national_poverty_states_br_T.reset_index(drop=False)
print(national_poverty_states_br_T)
```

```python
#Renomeando nosso index e removendo a primeira linha repetida
new_columns_names = national_poverty_states_br_T.iloc[0].tolist()
new_columns_names= dict(zip(national_poverty_states_br_T.columns, new_columns_names))
national_poverty_states_br_T.rename(columns = new_columns_names, inplace=True)
print(national_poverty_states_br_T)
```

```python
national_poverty_states_br_T = national_poverty_states_br_T.drop(index=0)
print(national_poverty_states_br_T)
```

```python
national_poverty_states_br_T.rename(columns = {'Código' : 'time'}, inplace = True)
print(national_poverty_states_br_T)
```

```python
#Substituindo as , por .
national_poverty_states_br_T = national_poverty_states_br_T.replace(',', '.', regex=True)
print(national_poverty_states_br_T)
```

```python
national_poverty_states_br_T[[11,12,13,14,15,16,17,21,22,23,24,25,26,27,28,29,31,32,33,35,41,42,43,50,51,52,53]] = national_poverty_states_br_T[[11,12,13,14,15,16,17,21,22,23,24,25,26,27,28,29,31,32,33,35,41,42,43,50,51,52,53]].astype(float)
print(national_poverty_states_br_T)
```

```python
national_poverty_states_br_T['time'] = national_poverty_states_br_T['time'].astype(int)
print(national_poverty_states_br_T)
```
