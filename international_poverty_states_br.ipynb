```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from bs4 import BeautifulSoup
import requests
import warnings
warnings.filterwarnings('ignore')
```

```python
#Carregando nosso DF
os.chdir('/content/drive/MyDrive/Bases de Dados para Indicadores EUROFOUND/Base de Dados')
international_poverty_states_br = pd.read_csv('taxa_pobreza_internacional.csv', sep =';')
```

```python
international_poverty_states_br.head()
```

```python
#Selecionando apenas as colunas que vamos precisar
international_poverty_states_br = international_poverty_states_br.drop(['Sigla', 'Estado'], axis = 1)
```

```python
international_poverty_states_br.head()
```

```python
#Transpondo nosso DF
international_poverty_states_br_T = international_poverty_states_br.T
international_poverty_states_br_T = international_poverty_states_br_T.set_index(international_poverty_states_br.columns)
```

```python
print(international_poverty_states_br_T)
```

```python
print(international_poverty_states_br_T.iloc[0])
```

```python
print(international_poverty_states_br_T)
```
